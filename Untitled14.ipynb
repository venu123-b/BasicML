Importing libraries


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Extract the dataset 

```
titanic=pd.read_csv('train.csv')
titanic.head(8)
```

Lineplot

```
titanic.isnull()
```

Lineplot

```
sns.lineplot(data=titanic,x="Age",y="Survived")
plt.show()
```

Heat map

```
sns.heatmap(titanic.isnull())
```

count plot

```
sns.set_style('darkgrid')
sns.countplot(x='Survived',hue='Sex',data=titanic,palette='BuGn')
```

```
sns.set_style('darkgrid')
sns.countplot(x='Survived',hue='Pclass',data=titanic,palette='cubehelix')
```

```
sns.countplot(x='SibSp',data=titanic,palette='Blues_r')
plt.show()
```

distplot

```
sns.distplot(titanic['Age'].dropna(),color='darkviolet',bins=10)
plt.show()
```

```
catplot
```

relplot

```
sns.catplot(x="Pclass",y="Age", data=titanic,kind="swarm")
plt.show()
```

relplot

```
sns.relplot(x="Pclass",y="Age", data=titanic,kind="line",hue="Sex")
plt.show()
```

boxplot

```
sns.catplot(x="Pclass",y="Age", data=titanic,kind="box")
plt.show()
```

boxenplot

```
sns.catplot(x="Pclass",y="Age", data=titanic,kind="boxen")
plt.show()
```

barplot

```
sns.barplot(x="Pclass",y="Age", data=titanic)
plt.show()
```
